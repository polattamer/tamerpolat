## Question 1

Design and implement and object oriented program for quaternions. Necessary information about the topic is given in appendix. Write your design and code on this notebook. (100 pts)

Design

> **Quaternion**

> *Features*
- a
- b
- c
- d

> *Method*
- add (+)
- sub (-)
- center 
- conjugation
- inverse
- hamilton product


```python

class Quaternion:
    def __init__(self, a, b, c, d):
        self.a = a
        self.b = b
        self.c = c  
        self.d = d
    def __str__(self):
        return "({}+{}i+{}j+{}k) " .format(self.a, self.b, self.c, self.d)

    def __add__(self, other):
        return Quaternion(self.a + other.a, self.b + other.b, self.c + other.c, self.d + other.d)
    
    def __sub__(self, other):
        return Quaternion(self.a - other.a, self.b - other.b, self.c - other.c, self.d - other.d)

    def __int__(self):
        return int(self.a)

    def conjugate(self):
        return "({}-{}i-{}j-{}k) " .format(self.a, self.b, self.c, self.d)
    
    def inverse(self):
        ters = (self.a**2+self.b**2+self.c**2+self.d**2)
        return "({}-{}i-{}j-{}k) " .format(self.a/ters, self.b/ters, self.c/ters, self.d/ters)
    
    def __mul__(self, other):
        hamilton1 = self.a*other.a - self.b*other.b - self.c*other.c - self.d*other.d
        hamilton2 = self.a*other.b + self.b*other.a + self.c*other.d - self.d*other.c
        hamilton3 = self.a*other.c - self.b*other.d + self.c*other.a + self.d*other.b
        hamilton4 = self.a*other.d + self.b*other.c - self.c*other.b + self.d*other.a
        return "({}+{}i+{}j+{}k) " .format(hamilton1, hamilton2, hamilton3, hamilton4)


my_qua = Quaternion (4,5,6,7)
my_qua1 = Quaternion (1,2,3,9)
#My First Quaternion
print("My first Quaternion element;")
print(my_qua)

#and my second Quaternion
print("and my second Quaternion element;")
print(my_qua1)

#Add 
print("Adding two elements each other;")
print((my_qua+my_qua1))

#Sub
print("subtraction of two elements;")
print((my_qua-my_qua1))

#Center of the Quaternions q = a 
print("Center of the my first Quaternion element;")
print(int(my_qua))

#Conjugation of Quaternions
print("Conjugation for first element;")
print(my_qua.conjugate())

#Inverse
print("Inverse;")
print(my_qua.inverse())

#Hamilton Product for two elements
print("Applying Hamilton product for two elements;")
print(my_qua*my_qua1)
```
